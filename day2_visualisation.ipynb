```
#przekazujemy średnią/medianę w sposób dynamiczny
#mediana jest bardziej odporna na wartości odstające
def group_and_barplot(feat_groupby, feat_agg = 'price_value', agg_funcs = [np.mean, np.median, np.size], feat_sort = 'mean', top = 100, subplots = True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending = False)
    .head(top)
  ).plot(kind = 'bar', figsize = (15, 5), subplots = subplots)
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot(['param_rodzaj-paliwa'], feat_sort = 'mean');
```
